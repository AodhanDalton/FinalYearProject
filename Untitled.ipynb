```python
import pandas as pd 
import numpy as np
import os
import os.path as op
import mne
from mne.io import concatenate_raws, read_raw_edf
from mne import io
```

# Data Acquisition

```python
import warnings; warnings.simplefilter('ignore')

data_files = os.listdir('Data-set/')

raw_files =  [mne.io.read_raw_gdf('Data-set/' + f, preload=True) for f in data_files]
```

# Data Pre-Processing

```python
raw = mne.io.concatenate_raws(raw_files)

fname = raw
evoked = mne.read_evokeds(fname, condition='Left Auditory',baseline=(None, 0))

# plot with bads
evoked.plot(exclude=[])

# compute interpolation (also works with Raw and Epochs objects)
evoked.interpolate_bads(reset_bads=False)

# plot interpolated (previous bads)
evoked.plot(exclude=[])
```
